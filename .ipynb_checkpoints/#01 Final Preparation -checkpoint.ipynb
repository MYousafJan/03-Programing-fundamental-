```python
num = int (input ("Enter an intager number:"))
reverse = 0
while num>0:
    digit= num%10
    reverse = reverse *10 +digit 
    num//=10
    print ("Rversed number:",reverse)
```

```python
# Program to reverse an integer using a loop

num = int(input("Enter an integer number: "))
reverse = 0

while num > 0:
    digit = num % 10          # Get the last digit	reverse = reverse * 10 + digit  # Build the reversed
number
    num //= 10                # Remove the last digit

print("Reversed number:", reverse)
```
