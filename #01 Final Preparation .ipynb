```python
# Program to calculate the occupancy rate of a hotel

# Input validation for number of floors
floors = int(input("Enter the number of floors in the hotel: "))
while floors < 1:
    print("Number of floors must be at least 1.")
    floors = int(input("Enter the number of floors in the hotel: "))

total_rooms = 0
total_occupied = 0

# Loop through each floor
for floor in range(1, floors + 1):
    print(f"\n--- Floor {floor} ---")
    rooms = int(input(f"Enter number of rooms on floor {floor}: "))	# Validate rooms per floor (must be 10 or more)
    while rooms < 10:
        print("Each floor must have at least 10 rooms.")
        rooms = int(input(f"Enter number of rooms on floor {floor}: "))

    occupied = int(input(f"Enter number of occupied rooms on floor {floor}: "))

    total_rooms += rooms
    total_occupied += occupied

# Calculate occupancy rate
occupancy_rate = (total_occupied / total_rooms) * 100

# Display results
print("\n===== Hotel Occupancy Report =====")
print("Total rooms in hotel:", total_rooms)
print("Total occupied rooms:", total_occupied)
print(f"Occupancy rate: {occupancy_rate:.2f}%")
```
